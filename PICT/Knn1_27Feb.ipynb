```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ingredients.csv')
df.head(1)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Visualize
sns.jointplot(x = 'sweetness', y = 'crunchiness',
             hue = 'class', data = df);
```

```python
# separate input and output
```

```python
x = df[['sweetness', 'crunchiness']]
y = df['class']
```

## Build the model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 4)
```

```python
knn.fit(x, y)
```

## prediction

```python
tomato = pd.DataFrame({
    'sweetness': [6],
    'crunchiness': [4]
})
```

```python
tomato
```

```python
knn.predict(tomato)
```
