```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mtcars.csv')
df
```

```python
df.columns
```

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x,y)
```

```python
lr.score(x, y)
```

### Recursive feature elimination

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(lr)
```

```python
x = df.drop('mpg', axis = 1)
y = df['mpg']
```

```python
rfe.fit(x, y)
```

```python
rfe.score(x, y)
```

```python
x.columns
```

```python
rfe.get_support()
```

```python
x = x.iloc[:, rfe.get_support()]
```

```python
x
```

### forward/ backward feature selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sfs = SequentialFeatureSelector(lr, direction= 'backward')
```

```python
x = df.drop('mpg', axis = 1)
```

```python
sfs.fit(x, y)
```

```python
sfs.get_support()
```

```python
x = x.iloc[:, sfs.get_support()]
```

```python
x
```

```python
lr.fit(x, y)
print(lr.score(x, y))
```
